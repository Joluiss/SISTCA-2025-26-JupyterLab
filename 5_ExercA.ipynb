## Lab script for introducing Python

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import random
import time
```

## Exercise A: Rock 🪨, Paper 📄, Scissors ✂️
### Game rules:
- Rock beats Scissors
- Paper beats Rock
- Scissor beats Paper

```python
RULES = {
    ("paper", "rock"):     "You win! 🎉",
    ("rock", "scissors"):  "You win! 🎉",
    ("scissors", "paper"): "You win! 🎉",
    ("rock", "paper"):     "Computer wins! 🤖",
    ("scissors", "rock"):  "Computer wins! 🤖",
    ("paper", "scissors"): "Computer wins! 🤖",
}
option = {"rock": "🪨", "paper": "📄", "scissors": "✂️"}
def get_result(user, computer):
    if user == computer:
        return "It's a tie! 🤝"
    return RULES[(user, computer)]

print("Pick your option against the Computer:")

buttons = {
    choice: widgets.Button(
        description=f"{option[choice]} {choice.capitalize()}",
        layout=widgets.Layout(width="130px", height="40px"),
        button_style="info"
    )
    for choice in ["rock", "paper", "scissors"]
}

quit_btn = widgets.Button(description="Quit 🚪", button_style="danger",
                          layout=widgets.Layout(width="100px", height="40px"))
output = widgets.Output()

def on_choice(b):
    user = b.description.split(" ", 1)[1].lower()
    computer = random.choice(list(option.keys()))
    result = get_result(user, computer)
    with output:
        clear_output()
        print(f"You:      {option[user]}  {user.capitalize()}")
        time.sleep(1)
        print(f"Computer: {option[computer]}  {computer.capitalize()}")
        print(f"\n{result}")

def on_quit(b):
    with output:
        clear_output()
    for btn in buttons.values():
        btn.disabled = True
    quit_btn.disabled = True

for btn in buttons.values():
    btn.on_click(on_choice)
quit_btn.on_click(on_quit)

display(widgets.HBox([*buttons.values(), quit_btn]), output)
```

## Exercise B: Shapes and Sizes 📐
### Objectives:
- choose the length and the width of a rectangle to obtain the area and perimeter of it, as well as if the rectangle has equal sides

```python
class Rectangle():
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def surface(self):
        surface = self.length*self.width
        return surface
    def perimeter(self):
        perimeter = self.length*2 + self.width*2
        return perimeter
    def rectangle_equalSides(self):
        return self.length == self.width


length_input = widgets.BoundedFloatText(description= "Length:", value=1, min=1.0)
width_input = widgets.BoundedFloatText(description= "Width:", value=1, min=1.0)
submit = widgets.Button(description = "Calculate", button_style ="success")
output = widgets.Output()

    

def on_submit(b):
    with output:
        clear_output()
        Inputr = Rectangle(length_input.value, width_input.value)    
        print(f"Surface:     {Inputr.surface()}")
        print(f"Perimeter:   {Inputr.perimeter()}")
        if Inputr.rectangle_equalSides():
            print(f"The rectangle has equal signs.")
        else:
            print(f"The rectangle doesn't have equal sides.")

submit.on_click(on_submit)
display(length_input, width_input, submit, output)
```
